```python
from pyspark.sql.functions import col, sum, rank, dense_rank, row_number
from pyspark.sql.window import Window

# 70
df_full = orders \
    .join(customers, "customer_id") \
    .join(order_items, "order_id") \
    .join(products, "product_id")

df_full = df_full.withColumn(
    "revenue",
    col("quantity") * col("product_price")
)


# 71. Rank products by price
w1 = Window.orderBy(col("product_price").desc())
products.withColumn("rank", rank().over(w1)).show()


# 72. Rank products within category
w2 = Window.partitionBy("category").orderBy(col("product_price").desc())
products.withColumn("rank", rank().over(w2)).show()


# 73. Row number for customers by signup_date
customers.withColumn(
    "row_num",
    row_number().over(Window.orderBy("signup_date"))
).show()


#  Create revenue per customer (needed for next)
cust_rev = df_full.groupBy("customer_name", "city") \
    .agg(sum("revenue").alias("total_revenue"))


# 74. Rank customers by revenue
cust_rev.withColumn(
    "rank",
    rank().over(Window.orderBy(col("total_revenue").desc()))
).show()

# 75. Dense rank customers by revenue
cust_rev.withColumn(
    "dense_rank",
    dense_rank().over(Window.orderBy(col("total_revenue").desc()))
).show()


# 76. Find top customer per city
cust_rev.withColumn(
    "rank",
    rank().over(Window.partitionBy("city").orderBy(col("total_revenue").desc()))
).filter(col("rank") == 1).show()


#  77. Find top product per category
prod_rev = df_full.groupBy("category", "product_name") \
    .agg(sum("revenue").alias("total_revenue"))

prod_rev.withColumn(
    "rank",
    rank().over(Window.partitionBy("category").orderBy(col("total_revenue").desc()))
).filter(col("rank") == 1).show()


#  78. Running total revenue by order date
df_full.withColumn(
    "running_total",
    sum("revenue").over(Window.orderBy("order_date"))
).show()


# 79. Running total revenue by product
df_full.withColumn(
    "running_total",
    sum("revenue").over(Window.partitionBy("product_id"))
).show()


# 80. Rank orders by payment amount
payments.withColumn(
    "rank",
    rank().over(Window.orderBy(col("amount").desc()))
).show()
```

```python
# Create temp views
customers.createOrReplaceTempView("customers")
orders.createOrReplaceTempView("orders")
products.createOrReplaceTempView("products")
order_items.createOrReplaceTempView("order_items")
payments.createOrReplaceTempView("payments")

# 91
spark.sql("SELECT * FROM customers").show()

# 92
spark.sql("SELECT * FROM customers WHERE age > 30").show()

# 93
spark.sql("SELECT city, COUNT(*) AS total FROM customers GROUP BY city").show()

# 94
spark.sql("SELECT city, AVG(age) AS avg_age FROM customers GROUP BY city").show()

# 95
spark.sql("""
SELECT c.customer_name, o.status
FROM customers c
JOIN orders o
ON c.customer_id = o.customer_id
""").show()

# 96
spark.sql("""
SELECT customer_id, COUNT(*) AS total_orders
FROM orders
GROUP BY customer_id
""").show()

# 97
spark.sql("""
SELECT *
FROM order_items oi
JOIN products p
ON oi.product_id = p.product_id
""").show()

# 98 (correct revenue calculation)
spark.sql("""
SELECT SUM(oi.quantity * p.product_price) AS total_revenue
FROM order_items oi
JOIN products p
ON oi.product_id = p.product_id
""").show()

# 99 (top 3 customers by revenue)
spark.sql("""
SELECT c.customer_name, SUM(oi.quantity * p.product_price) AS total_revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY c.customer_name
ORDER BY total_revenue DESC
LIMIT 3
""").show()

# 100 (top product by revenue)
spark.sql("""
SELECT p.product_name, SUM(oi.quantity * p.product_price) AS total_revenue
FROM order_items oi
JOIN products p ON oi.product_id = p.product_id
GROUP BY p.product_name
ORDER BY total_revenue DESC
LIMIT 1
""").show()
```

```python
from pyspark.sql.functions import col, sum, rank
from pyspark.sql.window import Window

# 1️⃣ Join all tables
final_df = orders \
    .join(customers, "customer_id") \
    .join(order_items, "order_id") \
    .join(products, "product_id") \
    .join(payments, "order_id")

# 2️⃣ Calculate revenue
final_df = final_df.withColumn(
    "revenue",
    col("quantity") * col("product_price")
)

# 3️⃣ Top customers
top_customers = final_df.groupBy("customer_name") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc())

top_customers.show()

# 4️⃣ Top products
top_products = final_df.groupBy("product_name") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc())

top_products.show()

# 5️⃣ Revenue per city
revenue_city = final_df.groupBy("city") \
    .agg(sum("revenue").alias("total_revenue"))

revenue_city.show()

# 6️⃣ Rank customers by revenue
window_spec = Window.orderBy(col("total_revenue").desc())

ranked_customers = top_customers.withColumn(
    "rank",
    rank().over(window_spec)
)

ranked_customers.show()

# 7️⃣ Save final dataset
final_df.write.mode("overwrite").csv("ecommerce_report", header=True)
```
