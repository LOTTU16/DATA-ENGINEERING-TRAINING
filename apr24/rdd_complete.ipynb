```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("RDD Basics").getOrCreate()
sc=spark.sparkContext
print("Spark Ready")
```

```python
numbers=[10,20,30,40,50]
rdd=sc.parallelize(numbers)
print(rdd)
```

```python
rdd.collect()
```

```python
data=["Rahul","Sneha","Arjun","Priya"]

name_rdd=sc.parallelize(data)
name_rdd.collect()
```

```python
rdd.glom().collect()
```

```python
rdd=sc.parallelize([1,2,3,4,5])
mapped_rdd=rdd.map(lambda x:x*2)
filtered_rdd=mapped_rdd.filter(lambda x:x>5)
print("Nothing executed yet")
```

```python
filtered_rdd.collect()
```

```python
rdd=sc.parallelize(["apple","banana","mango","kiwi"])
long_words=rdd.filter(lambda x:len(x)>5)
long_words.collect()
```

```python
lines=sc.parallelize([
    "python spark hadoop",
    "spark sql pyspark"
])
words=lines.flatMap(lambda line:line.split(" "))
words.collect()
```

```python
lines=sc.parallelize(["hello world","pyspark rdd"])
result=lines.map(lambda x:x.split(" "))
result.collect()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL Lab").getOrCreate()


employees_data = [
    (1, "Rahul", "IT", 70000, "Hyderabad"),
    (2, "Sneha", "HR", 60000, "Bangalore"),
    (3, "Arjun", "IT", 75000, "Chennai"),
    (4, "Priya", "Finance", 80000, "Hyderabad"),
    (5, "Karan", "IT", 50000, "Mumbai"),
    (6, "Amit", "HR", 58000, "Delhi"),
    (7, "Meera", "Finance", 82000, "Bangalore")
]

employees_cols = ["emp_id", "name", "department", "salary", "city"]

employees_df = spark.createDataFrame(employees_data, employees_cols)
employees_df.show()

departments_data = [
    ("IT", "Technology"),
    ("HR", "People Operations"),
    ("Finance", "Accounts and Finance")
]

departments_cols = ["department", "dept_full_name"]

departments_df = spark.createDataFrame(departments_data, departments_cols)
departments_df.show()

sales_data = [
    (101, 1, "Laptop", 1, 75000),
    (102, 2, "Mouse", 3, 500),
    (103, 3, "Keyboard", 2, 1500),
    (104, 1, "Monitor", 1, 12000),
    (105, 4, "Laptop", 1, 75000),
    (106, 3, "Mouse", 2, 500),
    (107, 5, "Keyboard", 1, 1500),
    (108, 1, "Laptop", 1, 75000)
]

sales_cols = ["sale_id", "emp_id", "product", "quantity", "price"]

sales_df = spark.createDataFrame(sales_data, sales_cols)
sales_df.show()
```

```python
employees_df.createOrReplaceTempView("employees")
departments_df.createOrReplaceTempView("departments")
sales_df.createOrReplaceTempView("sales")
```

```python
spark.sql("""
SELECT e.emp_id, e.name, e.department, d.dept_full_name
FROM employees e
INNER JOIN departments d
ON e.department = d.department
""").show()
```

```python
spark.sql("""
SELECT e.emp_id, e.name, e.department, d.dept_full_name
FROM employees e
LEFT JOIN departments d
ON e.department = d.department
""").show()
```

```python
spark.sql("""
SELECT
    COUNT(*) AS total_employees,
    SUM(salary) AS total_salary,
    AVG(salary) AS avg_salary,
    MAX(salary) AS max_salary,
    MIN(salary) AS min_salary
FROM employees
""").show()
```

```python
spark.sql("""
SELECT department, COUNT(*) AS emp_count
FROM employees
GROUP BY department
""").show()
```

```python
spark.sql("""
SELECT department, AVG(salary) AS avg_salary
FROM employees
GROUP BY department
""").show()
```

```python
spark.sql("""
SELECT e.name, SUM(s.quantity * s.price) AS total_sales
FROM employees e
JOIN sales s
ON e.emp_id = s.emp_id
GROUP BY e.name
""").show()
```

```python
spark.sql("""
SELECT department, AVG(salary) AS avg_salary
FROM employees
GROUP BY department
HAVING AVG(salary) > 65000
""").show()
```

```python
spark.sql("""
SELECT product, SUM(quantity) AS total_qty
FROM sales
GROUP BY product
HAVING SUM(quantity) > 2
""").show()
```

```python
spark.sql("""
SELECT *
FROM employees
ORDER BY salary DESC
""").show()
```

```python
spark.sql("""
SELECT product, SUM(quantity * price) AS revenue
FROM sales
GROUP BY product
ORDER BY revenue DESC
""").show()
```

```python
spark.sql("""
SELECT
    emp_id,
    name,
    department,
    salary,
    RANK() OVER(PARTITION BY department ORDER BY salary DESC) AS salary_rank
FROM employees
""").show()
```

```python
spark.sql("""
SELECT
    emp_id,
    name,
    department,
    salary,
    ROW_NUMBER() OVER(PARTITION BY department ORDER BY salary DESC) AS row_num
FROM employees
""").show()
```

```python
spark.sql("""
SELECT
    emp_id,
    name,
    department,
    salary,
    ROW_NUMBER() OVER(PARTITION BY department ORDER BY salary DESC) AS dense_rank
FROM employees
""").show()
```
