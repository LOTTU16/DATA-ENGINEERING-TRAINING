```python

from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

```python
# Logs
logs = [
"Rahul login","Sneha login","Arjun login","Rahul purchase","Priya login",
"Sneha logout","Rahul login","Karan login","Arjun purchase","Sneha login",
"Rahul logout","Amit login","Priya purchase","Karan purchase","Sneha login",
"Rahul purchase","Meera login","Arjun login","Sneha purchase","Rahul login"
]
logs_rdd = sc.parallelize(logs)

# Sales
sales = [
("Laptop",1,75000),("Mouse",3,500),("Keyboard",2,1500),("Laptop",1,75000),
("Monitor",1,12000),("Mouse",2,500),("Keyboard",1,1500),("Laptop",1,75000),
("Monitor",2,12000),("Mouse",3,500),("Laptop",1,75000),("Keyboard",2,1500),
("Monitor",1,12000),("Mouse",1,500),("Laptop",1,75000)
]
sales_rdd = sc.parallelize(sales)

# Employees
employees = [
("Rahul","IT",70000),("Sneha","HR",60000),("Arjun","IT",75000),
("Priya","Finance",80000),("Karan","IT",50000),("Amit","HR",58000),
("Meera","Finance",82000),("Ravi","IT",72000),("Neha","HR",61000),
("Vikram","Finance",90000)
]
emp_rdd = sc.parallelize(employees)
```

```python
print("Exercise 1:", logs_rdd.count())

print("\nExercise 2:", logs_rdd.take(5))

users = logs_rdd.map(lambda x: x.split()[0])
actions = logs_rdd.map(lambda x: x.split()[1])

print("\nExercise 3:", users.collect())
print("\nExercise 4:", actions.collect())

print("\nExercise 5:", users.distinct().collect())

login_count = logs_rdd.filter(lambda x: "login" in x).count()
print("\nExercise 6:", login_count)

print("\nExercise 7:")
for row in logs_rdd.filter(lambda x: "purchase" in x).collect():
    print(row)
```

```python
sentences = sc.parallelize([
"big data spark",
"spark hadoop hive",
"python spark hadoop",
"spark streaming kafka"
])

words = sentences.flatMap(lambda x: x.split())

print("\nExercise 8:", words.collect())
print("\nExercise 9:", words.distinct().collect())

print("\nExercise 10:")
word_count = words.map(lambda x: (x,1)).reduceByKey(lambda a,b: a+b)
for w,c in word_count.collect():
    print(f"{w} → {c}")
```

```python
print("\nExercise 11:")
revenue = sales_rdd.map(lambda x: (x[0], x[1]*x[2]))
for p,r in revenue.collect():
    print(f"{p} → {r}")

total_revenue = revenue.map(lambda x: x[1]).reduce(lambda a,b: a+b)
print("\nExercise 12:", total_revenue)

print("\nExercise 13:", sales_rdd.map(lambda x: x[0]).distinct().collect())

print("\nExercise 14:")
for row in sales_rdd.filter(lambda x: x[1] > 1).collect():
    print(row)

print("\nExercise 15:")
qty = sales_rdd.map(lambda x: (x[0], x[1])).reduceByKey(lambda a,b: a+b)
for p,q in qty.collect():
    print(f"{p} → {q}")

print("\nExercise 16:")
rev_prod = revenue.reduceByKey(lambda a,b: a+b)
for p,r in rev_prod.collect():
    print(f"{p} → {r}")

print("\nExercise 17:")
top = rev_prod.sortBy(lambda x: x[1], False).take(1)
for p,r in top:
    print(f"{p} → {r}")
```

```python
print("\nExercise 18:", emp_rdd.map(lambda x: x[0]).collect())

print("\nExercise 19:")
for row in emp_rdd.filter(lambda x: x[2] > 70000).collect():
    print(row)

print("\nExercise 20:", emp_rdd.map(lambda x: x[1]).distinct().collect())

print("\nExercise 21:")
dept_sal = emp_rdd.map(lambda x: (x[1], x[2])).reduceByKey(lambda a,b: a+b)
for d,s in dept_sal.collect():
    print(f"{d} → {s}")

print("\nExercise 22:")
avg = emp_rdd.map(lambda x: (x[1], (x[2],1))) \
    .reduceByKey(lambda a,b: (a[0]+b[0], a[1]+b[1])) \
    .map(lambda x: (x[0], x[1][0]/x[1][1]))
for d,a in avg.collect():
    print(f"{d} → {int(a)}")

print("\nExercise 23:")
for row in emp_rdd.sortBy(lambda x: x[2], False).take(1):
    print(row)
```

```python
print("\nExercise 24:")
count_emp = emp_rdd.map(lambda x: (x[1],1)).reduceByKey(lambda a,b: a+b)
for d,c in count_emp.collect():
    print(f"{d} → {c}")

print("\nExercise 25:")
for row in emp_rdd.sortBy(lambda x: x[2], False).collect():
    print(row)

print("\nExercise 26:")
for s in emp_rdd.map(lambda x: x[2]).takeOrdered(3, key=lambda x: -x):
    print(s)

print("\nExercise 27:")
for row in emp_rdd.map(lambda x: (x[1], x[0])).collect():
    print(row)

print("\nExercise 28:")
grouped = emp_rdd.map(lambda x: (x[1], x[0])).groupByKey()
for d,names in grouped.collect():
    print(f"{d} → {list(names)}")
```

```python
print("\nExercise 29:")
for d,s in dept_sal.sortBy(lambda x: x[1], False).take(1):
    print(f"{d} → {s}")

print("\nExercise 30:")
dept_stats = emp_rdd.map(lambda x: (x[1], (1,x[2]))) \
    .reduceByKey(lambda a,b: (a[0]+b[0], a[1]+b[1]))
for d,(c,s) in dept_stats.collect():
    print(f"{d} → ({c}, {s})")
```

```python
print("\nFINAL MINI PROJECT")

rev_prod = sales_rdd.map(lambda x: (x[0], x[1]*x[2])) \
                    .reduceByKey(lambda a,b: a+b)

top_selling = sales_rdd.map(lambda x: (x[0], x[1])) \
                       .reduceByKey(lambda a,b: a+b) \
                       .sortBy(lambda x: x[1], False)

highest_revenue = rev_prod.sortBy(lambda x: x[1], False)

more_than_3 = sales_rdd.map(lambda x: (x[0],1)) \
                       .reduceByKey(lambda a,b: a+b) \
                       .filter(lambda x: x[1] > 3)

sorted_rev = rev_prod.sortBy(lambda x: x[1], False)

print("\nProduct Revenue")
for p,r in sorted_rev.collect():
    print(f"{p} → {r}")

print("\nTop Selling Product:")
for p,q in top_selling.take(1):
    print(f"{p} → {q}")

print("\nHighest Revenue Product:")
for p,r in highest_revenue.take(1):
    print(f"{p} → {r}")

print("\nProducts sold more than 3 times:")
for p,c in more_than_3.collect():
    print(f"{p} → {c}")
```
