```python
from pyspark.sql import SparkSession
spark= SparkSession.builder\
    .appName("PySpark Basics")\
    .getOrCreate()
print("Spark Session Created")
```

```python
print(spark version)
```

```python
print(spark.version)
```

```python
data=[
    ("Rahul",25,"Hyderabad"),
    ("Sneha",28,"Bangalore"),
    ("Arjun",24,"Chennai"),
    ("Priya",27,"Hyderabad")
]
columns=["Name","Age","City"]
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df.printSchema ()
```

```python
df.select("Name").show()
```

```python
df.select("Name","City").show()
```

```python
df.filter(df.Age>25).show()
```

```python
df.groupBy("City").count().show()
```

```python
df.orderBy("Age").show()
```

```python
df.orderBy(df.Age.desc()).show()
```
