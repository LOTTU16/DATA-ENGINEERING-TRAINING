```python
emp_df = spark.read.csv("employees.csv", header=True, inferSchema=True)
dept_df = spark.read.csv("departments.csv", header=True, inferSchema=True)
sales_df = spark.read.csv("sales.csv", header=True, inferSchema=True)
json_df = spark.read.json("employees_nested.json")
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank,row_number

w = Window.partitionBy("department").orderBy(col("salary").desc())

emp_df.withColumn("rank",rank().over(w)).show()
emp_df.withColumn("row_num",row_number().over(w)).show()
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def cat(x):
    if x>50000: return "Premium"
    elif x>=20000: return "Medium"
    else: return "Small"

sales_df.withColumn("category", udf(cat,StringType())("amount")).show()
```

```python
from pyspark.sql.functions import col, avg, sum
```

```python
top_employee = emp_df.orderBy(col("salary").desc()).first()
```

```python
avg_salary = emp_df.select(avg("salary")).first()[0]
```

```python
dept_count = emp_df.groupBy("department").count().collect()
```

```python
total_sales = sales_df.select(sum("amount")).first()[0]
```

```python
best_sales_emp = sales_df.groupBy("emp_id") \
    .sum("amount") \
    .orderBy(col("sum(amount)").desc()) \
    .first()
```

```python
with open("final_report.txt", "w") as f:
    f.write("----- FINAL REPORT -----\n\n")

    f.write(f"Top Employee: {top_employee}\n")
    f.write(f"Average Salary: {avg_salary}\n")

    f.write("\nDepartment Employee Count:\n")
    for row in dept_count:
        f.write(f"{row['department']} : {row['count']}\n")

    f.write(f"\nTotal Sales: {total_sales}\n")
    f.write(f"Best Sales Employee: {best_sales_emp}\n")
```
