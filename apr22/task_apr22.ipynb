```python
with open("logins.txt") as f:
    logins = [line.strip() for line in f]

print("\n--- TXT ANALYSIS ---")

print("All Names:", logins)
print("Total Logins:", len(logins))
print("Unique Users:", set(logins))

# Count
counts = {}
for name in logins:
    counts[name] = counts.get(name, 0) + 1

print("\nLogin Count Dictionary:")
print(counts)

# Top 3 (EXPECTED FORMAT)
top3 = dict(sorted(counts.items(), key=lambda x: x[1], reverse=True)[:3])

print("\nTop 3 Users:")
for k, v in top3.items():
    print(f"{k} → {v}")

# Users > 4
print("\nUsers with >4 logins:")
for k, v in counts.items():
    if v > 4:
        print(f"{k} → {v}")
```

```python
import csv

employees = []
with open("employees.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        row["salary"] = int(row["salary"])
        row["emp_id"] = int(row["emp_id"])
        employees.append(row)

print("\n--- EMPLOYEE ANALYSIS ---")

print("Total Employees:", len(employees))

print("\nIT Employees:")
for e in employees:
    if e["department"] == "IT":
        print(e)

print("\nSalary > 75000:")
for e in employees:
    if e["salary"] > 75000:
        print(e)

avg = sum(e["salary"] for e in employees)/len(employees)
print("\nAverage Salary:", avg)

print("\nHighest Paid:", max(employees, key=lambda x: x["salary"]))
print("Lowest Paid:", min(employees, key=lambda x: x["salary"]))

# Dept count
dept = {}
for e in employees:
    d = e["department"]
    dept[d] = dept.get(d, 0) + 1

print("\nEmployees per Department:", dept)

# City count
city = {}
for e in employees:
    c = e["city"]
    city[c] = city.get(c, 0) + 1

print("\nEmployees per City:", city)
```

```python
sales = []
with open("sales.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        row["quantity"] = int(row["quantity"])
        row["price"] = int(row["price"])
        row["emp_id"] = int(row["emp_id"])
        row["revenue"] = row["quantity"] * row["price"]
        sales.append(row)

print("\n--- SALES ANALYSIS ---")

total_rev = sum(s["revenue"] for s in sales)
print("Total Revenue:", total_rev)

# Revenue per product (EXPECTED FORMAT)
prod_rev = {}
for s in sales:
    p = s["product"]
    prod_rev[p] = prod_rev.get(p, 0) + s["revenue"]

print("\nRevenue per Product:")
for p, v in prod_rev.items():
    print(f"{p} → {v}")

# Best product
best = max(prod_rev, key=prod_rev.get)
print("\nBest Selling Product:", best)
```

```python
with open("orders.json") as f:
    data = json.load(f)

orders = data["orders"]

print("\n--- JSON ANALYSIS ---")

print("Total Orders:", len(orders))

total = sum(o["amount"] for o in orders)
print("Total Sales Amount:", total)

cust = {}
for o in orders:
    c = o["customer"]
    cust[c] = cust.get(c, 0) + o["amount"]

print("\nSpending per Customer:")
for k, v in cust.items():
    print(f"{k} → {v}")

print("\nHighest Spender:", max(cust, key=cust.get))
```

```python
print("\n--- FINAL REPORT ---")

# Mapping
emp_map = {e["emp_id"]: e["name"] for e in employees}

emp_total = {}
for s in sales:
    name = emp_map[s["emp_id"]]
    emp_total[name] = emp_total.get(name, 0) + s["revenue"]

print("\nEmployee Revenue:")
for name, rev in sorted(emp_total.items(), key=lambda x: x[1], reverse=True):
    print(f"{name} → {rev}")

# Save CSV
with open("final_sales_report.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Employee", "Revenue"])
    for name, rev in sorted(emp_total.items(), key=lambda x: x[1], reverse=True):
        writer.writerow([name, rev])

print("\n✅ File saved: final_sales_report.csv")
```
